- **RUNTIME ERRORS** - valid code failed to execute (moderately easy to fix)
- **SEMANTIC ERRORS** - code executes but the logic is not correct (hard to fix)

### try except else
**SYNTAX**
```python
try:
    statment # this will run first
except error1:
        statment #t his will run if error1 occurs
except error2,error3:
        statment # this will run if error2,error3 occurs
except error4 as var:
        statment # this will run if error4 occurs with assigning variable var
except:
        statment # for all other error, this will run
else:
    statement # run if there is no error
```

### try except else finally

````python
try:
    print("hello")
except:
    print("hi")
else:
    print("how re u")
finally:
    print("I am good")
````

### User defined error

````python
class Error(Exception):
    pass
class zerodivision(Error):
    pass
try:
    i=int(input("enter number"))
    if i==0:
        raise zerodivision
except zerodivision:
    print("value can't be 0")
    print()
````
