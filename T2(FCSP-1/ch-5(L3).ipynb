## issubset($a \subset b$),issuperset($b \supset a$),isdisjoint($a \coprod b$)  

````python
a={1,2,3,4,5,6,7}
b={4,5,6}
````

````python
print(a.issubset(b))
print(b.issubset(a))
print(a.issuperset(b))
print(b.issuperset(a))
````

````python
print(a.isdisjoint(b))
````

# Lembda Function
- it is a oneliner function
- it does not return value; it return function
- it is the nameless function with multiple arguments returning one valid python expression
**syntax**

```python
lambda x,y : x+y
```

````python
a=lambda x,y:x**y
print(a(5,4))
````

````python
print(a(3))
````

````python
a=lambda x:'even' if x%2==0 else 'odd'
print(a(8))
````

````python
add=lambda x,y=15,z=24:x+y+z 
````

````python
print(add(1))
````

````python
print(add(0,2,3))
````

## map,filter and reduce
### map

````python
l=[10,40,56,27,13,15,70]
````

````python
double=list(map(lambda x:x*2,l))
print(double)
````

````python
l=[1,2,3,4,5]
print(list(map(lambda x:'બેકી' if x%2==0 else 'એકી',l)))
````

### Filter

````python
l=[10,40,56,27,13,15,70]
````

````python
divide=list(filter(lambda x:x%4==0,l))
print(divide)
````

````python
divide=list(map(lambda x:x%4==0,l))
print(divide)
````

````python
Fruit=['apple','banana','apricot']
````

````python
print(list(filter(lambda x: x.startswith('a'),Fruit)))
````

## WAP to find the sum of positive number and negative number using lambda function 

````python
pos=sum(list(filter(lambda x: x<0,[-4,-2,0,1,2,3])))
print(pos)
neg=sum(list(filter(lambda x: x>0,[-4,-2,0,1,2,3])))
print(neg)
````
