### Dictionary

- mutable
- an associative array/mapping.
- represented as key value pair.
- indexing has no meaning.
- slicing is not present.
- keys are unique and immutable.
- values can be duplicated.

### Dictionary Comprehension

### WAP to return squares of numbers from 0 to 5 in a dictionary format

```python
{i: i**2 for i in range(6)}
```

```python
{i: i**2 for i in range(6) if i%2==0}
```
