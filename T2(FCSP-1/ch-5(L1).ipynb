## WAP to extract list of even positive number from a given list.

```python
l=[2,3,4,5,6,7,8]
m=[]
for i in l:
    if i>0 and i%2==0:
        m.append(i)
print(m)
```

### List Comprehension

### Syntax

```python
[expression for iterator in iterable if<condition>]
```

```python
l = [i for i in[-4,-3,0,2,4,5,7] if  i%2==0 and i>0]
print(l)
```

```python
l = sum([i for i in[-4,-3,0,2,4,5,7] if  i%2==0 and i>0])
print(l)
```

## WAP to find the sum of element in a list without using sum method

```python
l = [2,3,4,5,6,7,8]
total=0
for i in l:
    total+=i
print(f"The total of all list is {total}")
```

## WAP to find sum of element in a list except 13 which is unlucky number also don't count number immediately after 13

```python
l=[13,1,2,4,13]
total=0
skip=False
for i in l:
    if skip:
        skip=False
        continue
    if i==13:
        skip=True
        continue
    total+=i
print(f"The Sum of except 13 number is : {total}")
        
```
