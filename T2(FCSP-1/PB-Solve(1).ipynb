## WAP to convert number to roman

```python
def int_roman(num):
    # The list MUST be in descending order for the logic to work correctly
    value_to_roman = [
        (1000, "M"),
        (900, "CM"),
        (500, "D"),
        (400, "CD"),
        (100, "C"),
        (90, "XC"),
        (50, "L"),
        (40, "XL"),
        (10, "X"),
        (9, "IX"),
        (5, "V"),
        (4, "IV"),
        (1, "I"),
    ]
    # Spelling corrected: romannumber
    romannumber = ""
    for value, roman in value_to_roman:
        while num >= value:
            romannumber += roman
            num -= value
    return romannumber

# Rest of the code remains the same
number = int(input("Enter the number : "))
if 1 <= number <= 3999:
    roman = int_roman(number)
    print(f"The roman number for {number} is {roman}")
else:
    print("Please Enter number from 1 to 3999")
```

## Roman to int

```python
def roman_int(roman):
    romandict={
        "I":1,
        "V":5,
        "X":10,
        "L":50,
        "C":100,
        "D":500,
        "M":1000,
        
    }
    total=0
    previous=0
    for char in reversed(roman):
        current=romandict[char]
        if current<previous:
            total-=current
        else:
            total+=current
        previous=current
    return total

romannumber=input("Enter the Roman : ")
integer=roman_int(romannumber)
print(f"The integer value of {romannumber} is {integer}")
```

## From a given array of integer,find the length of the longest consecutive element sequance

```python
def longest(nums):
    result=0
    nums=set(nums)
    for num in nums:
        if num-1 not in nums:
            curr=num
            length=1
            while curr+1 in nums:
                curr+=1
                length+=1
            result=max(result,length)
    return result
longest([-3,-2,-1,0,6,7,9,8,4,5])
```

## WAP to convert number to english word

```python
def convert_words(num):
    units=["",'one','two','three','four','five','six','seven','eight','nine']
    teens=["",'Eleven','twelve','thirteen','fourteen','fifteen','sixteen','seventeen','eighteen','nineteen']
    tens=["",'ten','twenty','thirty','forty','fifty','sixty','seventy','eighty','ninty']
    
    def covertbelow100(n):
        if n<10:
            return units[n]
        elif 10<n<20:
            return teens[n-10]
        else:
            return tens[n//10]+" "+units[n%10]
        
    if num==0:
        return "Zero"
    result=" "
    if num>=100:
        result+=units[num//100]+" hundred "
    result+=covertbelow100(num%100)
    return result.strip()
```

```python
def getsubstring(s):
    n=len(s)
    count=0
    for i in range(n):
        countzero=0
        countone=0
        counttwo=0
        
        for j in range(i,n):
            if s[j]=="0":
                countzero+=1
            elif s[j]=="1":
                countone+=1
            elif s[j]=="2":
                counttwo+=1
            if countzero==countone and countone==counttwo:
                count+=1
                print(s[i:j+1])
    return count
getsubstring("102100211")
```
