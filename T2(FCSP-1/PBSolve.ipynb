### WAP to shift a number

```python
shift=6
n=12345
s=str(n)
length=len(s)
if shift>length:
    print(s[::-1])
else:
    x=s[shift:]+s[:shift]
print(x)
```

```python
def passwordcheck(password):
    l,u,p,d=0,0,0,0
    if len(password)>=8:
        for i in password:
            if i.isupper():
                u+=1
            elif i.islower():
                l+=1
            elif i.isdigit():
                d+=1
            elif i in "@_$":
                p+=1
        if l>=1 and u>=1 and d>=1 and p>=1 and l+p+u+d == len(password):
            print("Password is valid")
        else:
            print("invalid password")
    else:
        print("Less then 8 element - invalid password")
    
            
```

```python
passwordcheck("Ram@123")
```

### ABCDEFGHIJKLMNOPQRSTUVWXYZ

```python
def encrypt(message,key):
    message=message.upper()
    alpha="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    result=""
    for letter in message:
        if letter in alpha:
            letterindex=(alpha.find(letter)+key)%len(alpha)
            result+=alpha[letterindex]
        else:
            result+=letter
    return result
```

```python
print(encrypt("LJEIT ENG",3))
print(encrypt("OMLHW HQJ",-3))
```

```python
def bal(s1,s2):
    if len(s1)!=len(s2):
        return False
    return sorted(s1)==sorted(s2)

s1="hello"
s2="olleh"

if bal(s1,s2):
    print("Balanced")
else:
    print("Not balanced")
```
