-  Write a python program to make a module which contain all the basic functions related to string and import that module in 
another file and use that fuctions with string given by user.

```python
%%writefile stringmodule.py
def stringlen(s):
    return len(s) - s.count(" ")
def reversestring(s):
    return s[::-1]
```

```python
from stringmodule import stringlen,reversestring
```

```python
stringlen("Hello world")
```

```python
reversestring("hello world")
```

### 498
` 
Write a python program to make a module named cal.py which contain all the basic functions related to calculator like 
addition, subtraction, multiplication, and division import that module in another file and use that functions with number 
inputs given by user.
`

```python
%%writefile cal.py
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
```

```python
a=int(input("Enter a number"))
b=int(input("Enter a number"))
```

```python
import cal
```

```python
print(cal.add(a,b))
print(cal.sub(a,b))
print(cal.mul(a,b))
print(cal.div(a,b))
```
