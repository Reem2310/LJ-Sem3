### 475 
Write a â€œpagerâ€ program. Your solution should prompt for a filename, and display the text file 25 lines at a time, pausing 
each time to ask the user to enter the word â€œcontinueâ€, in order to show the next 25 lines or enter the word â€œstopâ€ to close the file.

````python
def pager(filename):
    with open(filename,encoding='utf-8') as f:
        lines = f.readlines()
        total_lines=len(lines)
        start = 0
        end=min(25,total_lines)
        while start<total_lines:
            for i in range(start,end):
                print(lines[i],end=" ")
            user=input("Enter 25 lines")
            if user.lower()=="stop":
                break
            start+=25
            end=min(start+25,total_lines)
                
````

````python
filename = "alice.txt"
````

````python
pager(filename)
````
